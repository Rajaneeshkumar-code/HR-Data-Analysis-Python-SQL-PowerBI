```python
# Remove duplicated rows and make sure to reset the index after dropping duplicates
df = df.drop_duplicates().reset_index(drop=True)

# Verify
print("Shape after removing duplicates:", df.shape)
print("Remaining duplicate rows:", df.duplicated().sum())
```

```python
# Check columns for mixed data types and text formatting inconsistencies

summary = []

for col in df.columns:
    s = df[col].dropna()
    type_count = s.map(type).nunique()

    # Text formatting checks for object/string columns
    strip_changes = False
    case_changes = False
    if pd.api.types.is_object_dtype(df[col]) or pd.api.types.is_string_dtype(df[col]):
        as_str = s.astype(str)
        strip_changes = (as_str != as_str.str.strip()).any()
        case_changes = as_str.str.lower().nunique() < as_str.nunique()

    summary.append({
        "column": col,
        "dtype": df[col].dtype,
        "mixed_python_types": type_count > 1,
        "num_python_types": type_count,
        "leading_trailing_spaces": strip_changes,
        "case_variation": case_changes
    })

summary_df = pd.DataFrame(summary)

# Show only columns that may have formatting issues
display(summary_df[
    (summary_df["mixed_python_types"]) |
    (summary_df["leading_trailing_spaces"]) |
    (summary_df["case_variation"])
])
```

```python
issue_cols = summary_df.loc[
    summary_df["mixed_python_types"] |
    summary_df["leading_trailing_spaces"] |
    summary_df["case_variation"],
    "column"
].tolist()

print("Columns with formatting issues:")
print(issue_cols if issue_cols else "None")
```

```python
salary_col = "SalarySlab" if "SalarySlab" in df.columns else "MonthlyIncome"

print(f"Checking format for column: {salary_col}")
print("Data type:", df[salary_col].dtype)

if pd.api.types.is_numeric_dtype(df[salary_col]):
    print("\nNumeric summary:")
    print(df[salary_col].describe())
else:
    print("\nUnique values:")
    print(df[salary_col].dropna().astype(str).unique())
    print("\nValue counts:")
    print(df[salary_col].value_counts(dropna=False))

    # Check for leading/trailing spaces and case inconsistencies
    s = df[salary_col].dropna().astype(str)
    print("\nHas leading/trailing spaces:", (s != s.str.strip()).any())
    print("Has case variation:", s.str.lower().nunique() < s.nunique())
```

```python
print(df['SalarySlab'])
```

```python
salary_col = "SalarySlab" if "SalarySlab" in df.columns else "salarylab"

salary_text = (
    df[salary_col]
    .astype(str)
    .str.lower()
    .str.strip()
    .str.replace(r"(\d+)k", r"\g<1>000", regex=True)
)

parts = salary_text.str.split("-", n=1, expand=True)

df["salary_lower"] = parts[0].str.extract(r"(\d+)")[0]
df["salary_upper"] = parts[1].str.extract(r"(\d+)")[0]

upto_mask = salary_text.str.contains("upto", na=False)
df.loc[upto_mask, "salary_lower"] = 0
df.loc[upto_mask, "salary_upper"] = salary_text.loc[upto_mask].str.extract(r"(\d+)")[0]

df["salary_lower"] = pd.to_numeric(df["salary_lower"], errors="coerce")
df["salary_upper"] = pd.to_numeric(df["salary_upper"], errors="coerce")

display(df[[salary_col, "salary_lower", "salary_upper"]].head(10))
```

## Step 4: Define KPI

```python
total_employees = df.shape[0]
print("Total Employees:", total_employees)
```

```python
attrition_count = df.loc[df["Attrition"] == "Yes"].shape[0]
print("Attrition Count:", attrition_count)
```

```python
attrition_rate = (attrition_count / total_employees) * 100
print(f"Attrition Rate: {attrition_rate:.2f}%")
```

```python
# KPI: Average Salary
salary_col = "MonthlyIncome" if "MonthlyIncome" in df.columns else "Salary"

average_salary = df[salary_col].mean()
print(f"Average Salary: {average_salary:.2f}")
```

```python
# KPI: Average Satisfaction Score

satisfaction_cols = [col for col in [
    "Satisfaction_Score",
    "JobSatisfaction",
    "EnvironmentSatisfaction",
    "RelationshipSatisfaction",
    "WorkLifeBalance"
] if col in df.columns]

if "Satisfaction_Score" in df.columns:
    avg_satisfaction = df["Satisfaction_Score"].mean()
else:
    avg_satisfaction = df[satisfaction_cols].mean(axis=1).mean()

print(f"Average Satisfaction Score: {avg_satisfaction:.2f}")
```
